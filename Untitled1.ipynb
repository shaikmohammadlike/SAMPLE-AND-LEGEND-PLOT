```python
import matplotlib.pyplot as plt
#creating a plot
plt.plot([1,2,3,4],[10,20,25,30])
#customizing axis limitsand labels
plt.xlim(0,5)
plt.ylim(0,35)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sample plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
year=[1,2,3]
veg_rate=[2,4,6]
milk_rate=[1,3,5]
fruit_rate=[3,6,9]
#plotting the data
line1,=plt.plot(year,veg_rate)
line2,=plt.plot(year,milk_rate)
line3,=plt.plot(year,fruit_rate)
plt.legend([line1,line2,line3],['vegetables','milk','fruits'])
plt.title("legend")
plt.show()
print('successfully placed a legend on the axes')
```
